## Q1

```python
from pandas import Series, DataFrame
import pandas as pd
%pylab inline
```

```python
# read in gold.txt
gold = pd.read_csv('./gold.txt', sep=" ", header=None)
gold[['url','category']] = gold[0].str.split(expand=True)
gold = gold[['url','category']]
gold
```

```python
# read in labels.txt
labels = pd.read_csv('./labels.txt', sep=" ", header=None)
labels[['turk','url','category']] = labels[0].str.split(expand=True)
labels = labels[['turk','url','category']]
labels
```

# Q2

```python
# create the labels_on_gold dataframe
labels_on_gold = labels.merge(gold, on='url', suffixes = ['_label','_gold'])
labels_on_gold  
```

```python
def url_g(s):
    if s in gold['url'].values:
        return True
    else:
        return False
    
labels_unknown = labels[labels['url'].apply(url_g) == False]
labels_unknown
```

## Q3

```python
# groupby function
grps = labels_on_gold.groupby('turk')

# build a dataframe rater_goodness
rater_goodness = pd.DataFrame()
rater_goodness['number_of_ratings'] = labels_on_gold.groupby('turk').count()['url']

for key, group in grps:
    correctness = len(group[group['category_label'] == group['category_gold']])/len(group) 
    rater_goodness.loc[key,'avg_correctness'] = correctness
    
rater_goodness
```

## Q4

```python
c = rater_goodness['avg_correctness']
rater_goodness['odds'] = c/(1.001-c)
rater_goodness
```

## Q5

```python
rater_goodness[rater_goodness['number_of_ratings'] >= 20].sort_values\
(by = "avg_correctness", ascending = False).head(10)
```

## Q6

```python
# create a correlation matrix between the number and correctness
rater_goodness[['number_of_ratings','avg_correctness']].corr()

# There seems to be not much relatedness between the two variables.
# We will further run OLS regression to testify this result.
```

```python
# create a linear model and test accuracy
import statsmodels.api as sm
from patsy import dmatrices

y, X = dmatrices('number_of_ratings ~ avg_correctness', data=rater_goodness, return_type='dataframe')
model = sm.OLS(y, X)       # Set up the model
result = model.fit()       # Fit model (find the intercept and slopes)
print(result.summary())

# The 95% confidence interval of the coefficient of independent variable includes zero
# and is not statistically significant.
# There is no proved significant relationship between number and correctness
```

## Q7

```python
# calculate the first 25 percentile
quantile = rater_goodness['number_of_ratings'].quantile(.75)
rater_sort = rater_goodness.sort_values(by = 'number_of_ratings', ascending = False)
rater_q = rater_sort[rater_sort['number_of_ratings'] >= quantile]
rater_q
```

```python
def url_r(s):
    if s in rater_q.index:
        return True
    else:
        return False
    
unknown = labels_unknown[labels_unknown['turk'].apply(url_r) == True]
unknown['odds'] = rater_q.loc[unknown['turk']]['odds'].values
unknown
```

```python
overall_odds = unknown.groupby(['url','category'])[['odds']].agg('product')
overall_odds
```

## Q8

```python
# pivot the overall_odds table
pivot = pd.pivot_table(overall_odds,
                          index = 'url',
                          columns=['category'],
                          values = ['odds'],
                          aggfunc = max)

# transpose and find the top category
category_top = DataFrame(pivot['odds'].T.idxmax())
category_top.head(5)
```

```python
# find the max odds
odds_top = DataFrame(pivot['odds'].T.max())
odds_top.head(5)
```

```python
# merge dataframes
result_75 = category_top.merge(odds_top, left_on = 'url', right_on = 'url')

# change column names
result_75.columns = ['top category', 'top odds']

result_75
```

## Q9

```python
# repeat Q7 for first 75%
# calculate the first 75 percentile
quantile_25 = rater_goodness['number_of_ratings'].quantile(.25)
rater_sort25 = rater_goodness.sort_values(by = 'number_of_ratings', ascending = False)
rater_q25 = rater_sort25[rater_sort25['number_of_ratings'] >= quantile_25]

def url_r(s):
    if s in rater_q25.index:
        return True
    else:
        return False
    
unknown25 = labels_unknown[labels_unknown['turk'].apply(url_r) == True]
unknown25['odds'] = rater_q25.loc[unknown25['turk']]['odds'].values

overall_odds25 = unknown25.groupby(['url','category'])[['odds']].agg('product')
overall_odds25
```

```python
# Repeat Q8 with the first 75%
# pivot the overall_odds25 table
pivot25 = pd.pivot_table(overall_odds25,
                          index = 'url',
                          columns=['category'],
                          values = ['odds'],
                          aggfunc = max)

# transpose and find the top category
category_top25 = DataFrame(pivot25['odds'].T.idxmax())

# find the max odds
odds_top25 = DataFrame(pivot25['odds'].T.max())

# merge dataframes
result_25 = category_top25.merge(odds_top25, left_on = 'url', right_on = 'url')

# change column names
result_25.columns = ['top category', 'top odds']

result_25
```

```python
# create the final table by merging the two previous tables
final = result_75.merge(result_25, left_on = 'url',\
                        right_on = 'url',\
                        suffixes = ['_75', '_25'])
final
```

```python
compare = pd.crosstab(final['top category_75'], final['top category_25'])
compare
```
